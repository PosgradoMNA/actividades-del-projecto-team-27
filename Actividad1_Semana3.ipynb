**Parte 1: Fundamentos de bases de datos**

Fundamentos de bases de datos y para ciencia de datos.
Las bases de datos es un punto muy importante en la ciencia de datos, ya que es el punto de partida para que podamos hacer un análisis y posteriormente ejecutar los modelos de predicción que se puedan usar. Para esto existen diferentes bases de datos una de las primordiales que se debe tener conocimiento ya que en la mayoría de las empresas son usadas son “Relational Database Management Systems”. Sin embargo, las herramientas que son usadas para este tipo de base de datos son solo para los datos estructurados, los datos no estructurados están enfocados en otros sistemas y obtener información puede ser un poco más complicado.
En el data set que tengamos por analizar es importante corregir entradas erróneas, así como encontrar partes incompletas o que sean irrelevantes para el análisis, para eso es necesario modificar o eliminar todos esos datos previamente a nuestro modelo. Unos de los pasos a identificar en este proceso son, tipos de datos faltantes, técnicas para manejar los datos faltantes e Identificación de valores atípicos y datos ruidosos.


Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.

Un “Data warehouse” es un repositorio de información que puede ser analizado para hacer decisiones más informadas. Los datos corren en un “data warehouse” mediante sistemas que ayuda y dan soporte para actividades de “Business Intelligence”.  Centralizan y consolidan grandes cantidades de datos de diferentes fuentes y sus capacidades analíticas permiten derivar conocimiento valioso mediante el uso de ciencia de datos.
